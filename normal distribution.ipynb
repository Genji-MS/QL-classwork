```python
def percentage_in_range(min_, max_, list_):
    count = 0
    for item in list_:
        if item >= min_ and item <= max_:
            count +=1
    return (count/len(list_))*100
```

```python
print (percentage_in_range(50,70,listName))
```

```python
print (percentage_in_range(70,100,listName))
```

```python
#what is the percentage the grade is above 70% using CDF
from scipy.stats import norm
#survival function returns as below, the value and greater than
print (norm.sf(70, loc=60, scale= 10))
#or one minus CDF as CDF returns the value and less than
print (1-(norm.cdf(70,60,scale=10)) )

#FACTS 68% of the data would be between MEAN +/- STD
```

```python
import matplotlib.pyplot as plt

def gaussian_f(x, mu, sigma):
    f = 1/np.sqrt(2*np.pi*sigma**2)*np.exp(-(x-mu)**2/2/(sigma**2))
    return f

# define the range for our list, from minimum to it maximum with given space(100 here)
x = np.linspace(-5,5,100)
y = [gaussian_f(i,0,1) for i in x]

plt.plot(x,y)
plt.grid()
```
